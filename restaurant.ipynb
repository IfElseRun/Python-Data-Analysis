```python
from connection import MySQLConnection
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

new_connection = MySQLConnection()
```

```python
# Data overview
# SQL query of the Chinese restaurant with the highest number of records
q = ('''SELECT  count(`INSPECTION DATE`) as NUMBER_OF_VIOLATIONS, STR_TO_DATE(`INSPECTION DATE`,'%m/%d/%Y') as DATE_OF_INSPECTION
FROM nyc_restaurants.DOHMH_New_York_City_Restaurant_Inspection_Results 
where CAMIS = '41683816' 
group by DATE_OF_INSPECTION
ORDER BY DATE_OF_INSPECTION ASC LIMIT 999999;''')

# Convert the SQL query to Pandas DataFrame
r = pd.read_sql(q,  new_connection.con)
r.plot.bar(x='DATE_OF_INSPECTION', y='NUMBER_OF_VIOLATIONS',figsize = (15, 5))
```

```python
r.dt.hist(figsize = (15, 4))
plt.title("Histogram of the Scores", fontsize=15)
plt.xlabel('Score', fontsize = 12)
```
